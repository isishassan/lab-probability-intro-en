## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

```python
# YOUR CODE HERE
white = 10
red = 12
black = 8
total = white + red + black
selections = 3
```

```python
#Taking a white ball out.

p = white / total

p
```

```python
# Probability of taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.

p_white_first = white / total
p_black_second_given_white = black / (total - 1)

p = p_white_first * p_black_second_given_white
p
```

```python
#Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.

p_white_first = white / total
p_black_second = black / total

p = p_white_first * p_black_second

p
```

```python
import math

math.factorial(5)
```

```python
#Probability of taking 3 balls and them being all different colours. assuming no replacement

#combination probability

#step 1: see above, all my variables

#step 2: total factorial divided by (selections factorial x (1-selections) factorial)

total_f = math.factorial(total)

selections_f = math.factorial(selections) 

non_selections_f = math.factorial(total-selections)

total_possible = total_f / (selections_f * non_selections_f)

total_possible

#step 3: favourable selections

favourable_1 = white * black * red          #this is not really the correct way to calculate this, ask ana. it may work coincidentally with no replacement

#comb approach
p_white = math.comb(white,1)

p_red = math.comb(red,1)

p_black = math.comb(black,1)

favourable_2 = p_white * p_red * p_black


#step 4: outcome
probability_1 = favourable_1 / total_possible

probability_1

probability_2 = favourable_2 / total_possible

probability_2
```

## Challenge 4

 * If someone flies to China, the probability of getting Covid is 1%  
 * If someone flies to Italy, the probability of getting Covid is 20%  

I roll 1 dice with numbers 1, 2, 3, 4, 5, 6:  
 * If 1: I fly to Italy
 * If not 1: I fly to China

During my trip, I got Covid. What is the probability that I flew to China?

```python
# YOUR CODE HERE

p_italy = 1/6
p_china = 5/6
ch_cov = 0.01
it_cov = 0.2

p_covid = (p_italy * it_cov + p_china * ch_cov)

p_covid

#bayes theorem. p(china) given p(covid) already happened

nominator = p_china * ch_cov
denominator = p_covid

p = nominator / denominator

p
```
